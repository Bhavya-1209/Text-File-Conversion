**Compare the performance (time taken) of C, C++, Java, R, and Python programs for coverting 1GB, 2GB, 3B, 4GB, and 5GB text files to upper case.**

```python
import matplotlib.pyplot as plt

# Sample data
x = [200,400,600,800,1000]
y1 = [48, 58, 68, 78, 88]     # C
y2 = [50, 62, 73, 83, 95]     # C++
y3 = [53, 64, 76, 88, 100]    # Java
y4 = [58, 72, 90, 115, 145]   # R
y5 = [63, 82, 102, 130, 180]  # Python

# Plotting
plt.plot(x, y1, label='C', linewidth=2)
plt.plot(x, y2, label='C++', linewidth=1.5)
plt.plot(x, y3, label='Java', linewidth=2.5)
plt.plot(x, y4, label='R', linewidth=2)
plt.plot(x, y5, label='Python', linewidth=2)

# Labeling axes and title
plt.xlabel('File Size (GB)')
plt.ylabel('Time Taken (sec)')
plt.title('Performance of Different Programming Languages')
plt.legend()

# Display the plot
plt.show()
```
